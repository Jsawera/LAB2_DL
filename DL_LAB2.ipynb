```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
# Task 1: Step Function

def step_scalar(x):
    return 1 if x >= 0 else 0

def step_tensor(x):
    return tf.cast(x >= 0, dtype=tf.int32)

input_step = tf.constant([-3.0, -1.0, 0.0, 2.0, 5.0])
output_step = step_tensor(input_step)
print("Step Function Output:", output_step.numpy())

x_vals = tf.linspace(-5.0, 5.0, 100)
plt.plot(x_vals, step_tensor(x_vals))
plt.title("Step Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task 2: ReLU

def relu_scalar(x):
    return max(0, x)

def relu_tensor(x):
    return tf.maximum(0.0, x)

input_relu = tf.constant([-4.0, -2.0, 0.0, 3.0, 5.0])
output_relu = relu_tensor(input_relu)
print("ReLU Output:", output_relu.numpy())

plt.plot(x_vals, relu_tensor(x_vals))
plt.title("ReLU Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task 3: Sigmoid

def sigmoid_scalar(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_tensor(x):
    return 1 / (1 + tf.exp(-x))

input_sigmoid = tf.constant([-2.0, 0.0, 1.0, 3.0])
output_sigmoid = sigmoid_tensor(input_sigmoid)
print("Sigmoid Output:", output_sigmoid.numpy())

plt.plot(x_vals, sigmoid_tensor(x_vals))
plt.title("Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task 4: Tanh

def tanh_scalar(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

def tanh_tensor(x):
    return tf.tanh(x)

input_tanh = tf.constant([-1.5, 0.0, 1.0, 2.0])
output_tanh = tanh_tensor(input_tanh)
print("Tanh Output:", output_tanh.numpy())

plt.plot(x_vals, tanh_tensor(x_vals))
plt.title("Tanh Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
# Task 5: Leaky ReLU

def leaky_relu_scalar(x, alpha=0.1):
    return x if x >= 0 else alpha * x

def leaky_relu_tensor(x, alpha=0.1):
    return tf.where(x >= 0, x, alpha * x)

input_leaky = tf.constant([-3.0, -1.0, 0.0, 2.0])
output_leaky_01 = leaky_relu_tensor(input_leaky, alpha=0.1)
output_leaky_03 = leaky_relu_tensor(input_leaky, alpha=0.3)
print("Leaky ReLU (alpha=0.1):", output_leaky_01.numpy())
print("Leaky ReLU (alpha=0.3):", output_leaky_03.numpy())

plt.plot(x_vals, leaky_relu_tensor(x_vals, alpha=0.1), label="alpha=0.1")
plt.plot(x_vals, leaky_relu_tensor(x_vals, alpha=0.3), label="alpha=0.3")
plt.title("Leaky ReLU Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 6: Softmax

def softmax_tensor(x):
    e_x = tf.exp(x - tf.reduce_max(x))  # for numerical stability
    return e_x / tf.reduce_sum(e_x)

input_softmax = tf.constant([1.0, 2.0, 3.0])
output_softmax = softmax_tensor(input_softmax)
print("Softmax Output:", output_softmax.numpy())
print("Sum of Softmax Output:", np.sum(output_softmax.numpy()))

plt.bar(range(len(input_softmax)), output_softmax.numpy())
plt.title("Softmax Function")
plt.xlabel("Class Index")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
